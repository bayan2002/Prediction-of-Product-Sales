```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fpath = "/content/drive/MyDrive/AXSOSACADEMY/01-Fundamentals/Week02/Data/sales_predictions_2023.csv"
df = pd.read_csv(fpath)
df.head()
```

```python
fig, ax = plt.subplots();
sns.histplot(data=df, x='Item_Outlet_Sales')
ax.set(title='Distribution of Item Outlet Sales', xlabel='Sales', ylabel='Count');
fig.tight_layout()
fig.savefig('hist_sales_distribution.png', dpi=300, facecolor='white');
```

```python
fig, ax = plt.subplots()
sns.histplot(data=df, x='Item_Outlet_Sales', hue='Outlet_Type', ax=ax)
ax.set(title='Distribution of Item Outlet Sales By Outlet Type', xlabel='Sales', ylabel='Count')
fig.tight_layout()
fig.savefig('hist_sales_by_outlet_type.png', dpi=300, facecolor='white');
```

```python
fig, ax = plt.subplots()
sns.boxplot(data=df, x='Item_Outlet_Sales', ax=ax)
ax.set_title('Statistical Summary of Sales and Outliers')
fig.tight_layout()
fig.savefig('boxplot_sales_outliers.png', dpi=300, facecolor='white');
```

```python
corr = df.corr(numeric_only=True)
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
ax.set_title('Correlation Heatmap of Numerical Features');
fig.tight_layout()
fig.savefig('heatmap_correlations.png', dpi=300, facecolor='white');
```

```python
val_counts = df['Item_Type'].value_counts()
fig, ax = plt.subplots(figsize=(9, 6))
sns.countplot(data=df, x='Item_Type', order=val_counts.index)
ax.tick_params(axis='x', rotation=45)
ax.set_title('Frequency of Each Item Type');
fig.tight_layout()
fig.savefig('countplot_item_types.png', dpi=300, facecolor='white');
```
