# Data Cleaning

```python
#Number of rows and colums
df_sales_predictions.shape
```

```python
df_sales_predictions.dtypes
```

```python
df_sales_predictions.duplicated().sum()
```

```python
df_sales_predictions = df_sales_predictions.drop_duplicates()
```

```python
df_sales_predictions.isna().sum()
```

```python
cat_cols = df_sales_predictions.select_dtypes('object').columns
cat_cols
df_sales_predictions[cat_cols].isna().sum()
df_sales_predictions[cat_cols] = df_sales_predictions[cat_cols].fillna("MISSING")
df_sales_predictions[cat_cols].isna().sum()
```

```python
num_cols = df_sales_predictions.select_dtypes('number').columns
num_cols
df_sales_predictions[num_cols].isna().sum()
df_sales_predictions[num_cols] = df_sales_predictions[num_cols].fillna(-1)
df_sales_predictions[num_cols].isna().sum()
```

```python
df_sales_predictions.isna().sum()
```

```python
cat_cols
for col in cat_cols:
  print(f"Value Counts for {col}")
  print(df_sales_predictions[col].value_counts())
  # Increasing readability by adding an empty line
  print('\n')
```
